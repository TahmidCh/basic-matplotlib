```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace (0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional method
```

```python
plt.plot(x,y)
#plt.show()
plt.xlabel('x Label')
plt.ylabel('y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented Method
```

```python
fig = plt.figure()

axes = fig.add_axes([0.9, 0.5, .8, .8])
#left, bottom, widhth and height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1= fig.add_axes([0.1, 0.1, .8, .8])
axes2= fig.add_axes([.2, .5, .8, .8])

axes1.plot(x,y)
axes1.set_title('Lower plot')

axes2.plot(y,x)
axes2.set_title('Upper Plot')
```

```python
fig,axes= plt.subplots(nrows=1, ncols=2)

#axes.plot(x,y)

#plt.tight_layout() #fixing the overlaps

#for current_ax in axes:
#    current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

# Figure size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows =2, ncols =1, figsize=(8,2))

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

# Saving of figure

```python
fig= plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label="X square")
ax.plot(x, x**3, label="X Cubed")

ax.legend()
```

# Plot Apprerance

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x,y, color='orange', lw= 3, alpha=0.9, ls=':',marker='o',markersize=10,
       markerfacecolor='yellow',markeredgewidth=2 ,markeredgecolor='black')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
